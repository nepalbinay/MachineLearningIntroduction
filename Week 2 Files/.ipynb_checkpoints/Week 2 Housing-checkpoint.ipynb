# Week 2 Housing Analysis

Importing package called 'zipfile' to unzip the .zip data

```python
import zipfile as zf
```

```python
files = zf.ZipFile('home_data.sframe.zip','r')
files.extractall()
files.close()
```

```python
import turicreate as tc
```

```python
sales = tc.SFrame('home_data.sframe')
```

```python
sales.head()
```

## Exploring the data for housing sales

```python
sales.column_types
```

```python
#tc.visualization.set_target('browser') To open it in browser
tc.visualization.set_target('auto')
tc.show(sales[1:5000]['sqft_lot'],sales[1:5000]['price'])
```

```python
tc.visualization.scatter(sales[1:5000]['sqft_lot'],sales[1:5000]['price'])
```

```python
tc.visualization.scatter(sales['sqft_living'], sales['price'], 
                         xlabel = 'Price of Houses', 
                         ylabel = 'Living Area (in sqft)',
                         title = 'Living Area Vs Price')
```

# Create a Simple Regression of sqft_living to Price

```python
train_data, test_data = sales.random_split(0.8, seed = 0)
```

## Build Regression Model

```python
sqft_model = tc.linear_regression.create(train_data, 
                                         target = 'price', 
                                         features = ['sqft_living'])
```

## Evaluate a Simple Model

```python
print(test_data['price'].mean())
```

```python
print(sqft_model.evaluate(test_data))
```

## Let's show what our prediction looks like

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(sales['sqft_living'], sales['price'])
```

```python
plt.plot(test_data['sqft_living'], test_data['price'],'.',
        test_data['sqft_living'], sqft_model.predict(test_data),'-')
```

```python
sqft_model.coefficients
```

```python
my_features = ['bedrooms','bathrooms','sqft_living','sqft_lot','floors','zipcode']
sales[my_features].show()
```

```python
tc.visualization.set_target('auto')
tc.visualization.box_plot(x = sales['zipcode'], y = sales['price'])
```

## Build a Regression Model with more features

```python
my_features_model = tc.linear_regression.create(train_data, target = 'price', features = my_features)
```

```python
print(sqft_model.evaluate(test_data))
print(my_features_model.evaluate(test_data))
```
